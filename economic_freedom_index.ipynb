```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
#Import CSV
csv_file = "Resources/economic_freedom_index_2019.csv"
econ_df = pd.read_csv(csv_file)
econ_df
```

```python
econ_df.head()
```

```python
# drop columns

columns = ['Country Name', 'World Rank','Population (Millions)', 'GDP (Billions, PPP)','Inflation (%)', 'Unemployment (%)']

econ_df_2 = econ_df[columns].copy()
econ_df_2
```

```python
# rename columns

econ_df_3 = econ_df_2.rename(columns={"Country Name": "country_name",
                                "World Rank": "world_rank",
                                "Population (Millions)": "population",
                                "GDP (Billions, PPP)": "total_gdp",
                                "Inflation (%)": "inflation",
                                "Unemployment (%)":"unemployment",
                                    
                                                                               })
econ_df_3.head()
```

```python
# set index
econ_df_3.set_index("country_name", inplace=True)
```

```python
econ_df_3
```

```python
# get rid of $ and ,
econ_df_3['total_gdp'] = econ_df_3['total_gdp'].str.replace(',', '').str.replace('$', '').astype(float)
```

```python
#econ_df.replace({"40.0 (2015 est.)":40.0}, inplace=True)
```

```python
#write to csv
econ_df_3.to_csv("Resources/economics.csv", index=True, header=True)
```
